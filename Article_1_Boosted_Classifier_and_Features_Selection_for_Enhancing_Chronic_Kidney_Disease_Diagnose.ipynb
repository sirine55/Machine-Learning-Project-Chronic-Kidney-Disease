# **Part 1: Business Understanding**




**Definition of CHRONIC KIDNEY DISEASES (CKD)** 

Your kidneys, each is just like the size of a computer mouse, they filter all the blood in your body every 30 minutes. 

They work hard to remove wastes, toxins, and excess fluid. 
They also help control blood pressure, stimulate production of red blood cells, keep your bones healthy and regulate blood chemicals that are essential to life.

CKD Chronic kidney disease, also called chronic kidney failure, involves a gradual loss of kidney function.
Advanced chronic kidney disease can cause dangerous levels of fluid, electrolytes and wastes to build up in your body .
If the patient fails to take drastic measures to improve his/her condition, both kidneys will fail, leading to acute renal failure.

When the symptoms become more severe,the disease can only be treated with artificial filtering (dialysis) and transplantation. 



**CKD by the Numbers :**



*   Kidney diseases are a leading cause of death in the United States.

*   About 37 million US adults are estimated to have CKD, and most are undiagnosed.

*  40% of people with severely reduced kidney function (not on dialysis) are not aware of having CKD.

* some countries have reached 200 cases/million per year and some other countries such as usa and taiwan have reached 400 cases/million 

* Every 24 hours, 360 people begin dialysis treatment for kidney failure.
*  In the United States, diabetes and high blood pressure are the leading causes of kidney failure, accounting for 3 out of 4 new cases.

*   In 2016, chronic kidney disease caused the death of 753 million people worldwide, where the number of males died was 336 million, while the number of females died was 417 million












**Risk factors for CKD :**

Some of these risk factors are  hypertension, diabetes, older age, cardiovascular disease (CVD), anemia. If a patient has at least one of  these factors he is  more likely to be diagnosed with CKD.


 **Symptoms of CKD :**



*   hypertension, or high blood pressure

*   hypotension or low blood pressure 

*   anemia
*  edema, or swollen feet, hands, and ankles

*   fatigue, or tiredness
*   bloody urine, in some cases


*   decreased urine output


*   decreased mental alertness, when the condition is severe



*  a loss of appetite


*   persistent itchy skin, when the condition is severe

*   more frequent urination, especially at night, in some cases

*  bone problems





**Side Effects :**

 CKD can cause damage to the nerve and immune system it can also cause hypertension and diabetes which are very serious. In 2012, diabetes and hypertension were categorized among the world’s top 10 causes of death. Diabetes was ranked ninth and killed 1.5 million people,Hypertension was ranked tenth and caused 1.1 million deaths.

**Stages :**

Medical experts determine kidney disease through glomerular filtration rate (GFR), which describes kidney function and it shows how well a person’s kidney is filtering waste . 
GFR is based on information such as body  size, age, blood test, gender, and other factors suffered by the patient. 
Regarding the GFR value, doctors can classify CKD into 5 stages.

CKD can be determined:
* kidneys damage(albuminuria )
* degradation of renal function glomerular filter rate (GFR) .



ckd.jpg

CKD can be prevented by early detection and proper medical handling. 
However, due to the increasing number of chronic kidney patients, the scarcity of specialist physicians, and the high costs of diagnosis and treatment, especially in developing countries, not all patients with chronic kidney disease get a proper diagnosis.


 Therefore, there is a need for computer-assisted diagnostics  of CKD is necessary to assist physicians and radiologists in supporting their diagnostic decisions. Technology like computer-aided diagnosis (CAD) can help and assist health care providers in diagnosing chronic kidney disease.

# **Part 3: Data Preprocessing**




Import the important libraries

```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.io.arff import loadarff 
```

Reading the data

```python
data = []
with open('chronic_kidney_disease.arff', "r") as f:
    for line in f:
        line = line.replace(' ', '')
        line = line.replace(',\n', '')
        line = line.replace('\n', '')
        line = line.replace('\t', '') 
        line = line.replace(',,', ',')
        data.append(line.split(','))


names = ['age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba',
         'bgr', 'bu',  'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc',
         'rbcc', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane',
         'class']
    
df = pd.DataFrame(data[29:], columns=names)
```

```python
#Now let's draw the new columns
sns.kdeplot(df_0["age"],color='blue',label='median')
sns.kdeplot(df_1["age"],color='orange',label='mean')
sns.kdeplot(df['age'],color='black',label='Original')
plt.legend()
#We have used 2 techniques to imputate the missing values and we noticed 
#that the mean imputation has the best compatible distribution curve on the original distribution curve.
#Therefore, it is best suited imputation technique for the column age
df["age"]=df_1["age"]
```

## Feature Selection 


Feature selection is a process where you automatically select those features in your data that contribute most to the prediction variable or output in which you are interested.

Three benefits of performing feature selection before modeling your data are:
*   Reduces Overfitting: Less redundant data means less opportunity to make decisions based on noise.
*   Improves Accuracy: Less misleading data means modeling accuracy improves
*   Reduces Training Time: Less data means that algorithms train faster.







**A- Filter Based** : We specify some metric and based on that filter features. An example of such a metric could be correlation/chi-square.






**1.   Correlation method by choosing the correlation threshold**




```python
#Using Pearson Correlation
plt.figure(figsize=(22,20))
cor = df.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.Reds)
plt.show()
```

```python
X=df.drop("class",axis=1)
y=df["class"]
```

```python
#Correlation with output variable
def cor_selector1(X, y,threshold):
  cor = df.corr()
  cor_target = abs(cor["class"])
  #Selecting highly correlated features , we choose the threshold=0.3
  relevant_features = cor_target[cor_target>threshold]
  return relevant_features.index.drop("class")

print(str((cor_selector1(X, y,0.3))), 'selected features')
```

**2. Correlation method by choosing the number of features to keep**



```python
X=df.drop('class',axis=1)
y=df['class']
num_feats=10
```

```python
def cor_selector2(X, y,num_feats):
    cor_list = []
    feature_name = X.columns.tolist()
    # calculate the correlation with y for each feature
    for i in X.columns.tolist():
        cor = np.corrcoef(X[i], y)[0, 1]
        cor_list.append(cor)
    # replace NaN with 0
    cor_list = [0 if np.isnan(i) else i for i in cor_list]
    # feature name
    cor_feature = X.iloc[:,np.argsort(np.abs(cor_list))[-num_feats:]].columns.tolist()

    return cor_feature

cor_feature = cor_selector2(X,y,num_feats)
print(str((cor_feature)), 'selected features')
```

### **Third Method : AdaBoost Classifier + CFS**



**How does the AdaBoost algorithm work?**

It works in the following steps:

1. Initially, Adaboost selects a training subset randomly.
It iteratively trains the AdaBoost machine learning model by selecting the training set based on the accurate prediction of the last training.
2. It assigns the higher weight to wrong classified observations so that in the next iteration these observations will get the high probability for classification.
3. Also, It assigns the weight to the trained classifier in each iteration according to the accuracy of the classifier. The more accurate classifier will get high weight.
4. This process iterate until the complete training data fits without any error or until reached to the specified maximum number of estimators.
5. To classify, perform a "vote" across all of the learning algorithms you built.
